```julia
using NBInclude
using Dates
using Plots
using DelimitedFiles
using Statistics
```

```julia
#Filenames
#Input
myfilename1="./Data/outExact.bin";    # Exact solution
myfilename2="./Data/outFCIRKh3.bin";  # FCIRK-LQ


#Information of the experiment
experiment1="./Data/experimentExact.txt"
experiment2="./Data/experimentFCIRKh3.txt";
```

```julia
planetslabels=["Sun" "Mercury" "Venus" "Mars" "Jupyter" "Saturn" "Uranus" "Neptune" "Pluto" "Ceres" "Pallas" "Vesta" "Iris" "Bamberga" "Earth" "Moon" ];
planetslabelsH=planetslabels[2:end];
```

## Import-data

```julia
f1=open(experiment1,"r");
expdata=readdlm(f1);
t0,tend,h1,nout1,neq=expdata
nout1=Integer(nout1)  # convert to Integer
neq=Integer(neq)      # convert to Integer
close(f1)
(h1,nout1,tend)
```

```julia
f1=open(experiment2,"r");
expdata=readdlm(f1);
t0,tend,h2,nout2,neq=expdata
nout2=Integer(nout2)  # convert to Integer
neq=Integer(neq)      # convert to Integer
close(f1)
(h2,nout2,tend)
```

```julia
t1,u1=Readbin(nout1,neq,myfilename1);
t2,u2=Readbin(nout2,neq,myfilename2);
ti=t1;
nout=nout2;
```

```julia
# Check if they are same initial conditions
norm(u1[1]-u2[1])
```

```julia
(t0,tend,nout,size(t1),neq)
```

## Get-Barycentric-Coordinates

Convert data from Heliocentric canonical coordinates to Barycentric-Coordinates 

```julia
(nbody,u0,k,Gm,iE,iM)= N16_Barycentric()
dim=3
neq=nbody*2*dim
nbodyH=nbody-1;
```

```julia
qq1,vv1,out1=GetBarycentric(u1,Gm,ti,nbodyH)
qq2,vv2,out2=GetBarycentric(u2,Gm,ti,nbodyH);
```

<a href="#top">Back to the top</a>

## Energy-error

```julia
# Initial Energy
a0=Array{BigFloat}(undef,neq+1);
a0=out2[1];
size(a0);
NBodyHam(neq,a0,Gm)
```

```julia
nstat=1;
HAMILTONIAN=NBodyHam;
(MeanHam1, DesvHam1)=EnergyError([out1],nstat,nout,HAMILTONIAN,neq,Gm);
(MeanHam2, DesvHam2)=EnergyError([out2],nstat,nout,HAMILTONIAN,neq,Gm);
```

```julia
plot(ti, log10.(abs.(MeanHam2)), label="FCIRK-LQ")
plot!(ti,log10.(abs.(MeanHam1)), label="FCIRK-Q")
plot!(title=MyTitle,
      xlabel="t",
      ylabel="Energy Error(log)",
      xlims=(t0,tend),
      ylims=(Eaxa,Eaxb),
      xtickfont = font(10, "Courier"),
      ytickfont = font(10, "Courier")
)
```

```julia
savefig("./Figures/Energy-FCIRK.pdf")
```

## Eccentricity-and-Semi-Major-Axis-error

```julia
mu=copy(k)
sun=parse(BigFloat,"0.295912208285591100e-3")
earth=parse(BigFloat,"0.888769244512563400e-9")
moon=parse(BigFloat,"0.109318945074237400e-10")
mu[end-1]=(earth+moon)+sun 
mu[end]=moon+earth
mu[end]=earth^3/(earth+moon)^2
axis1=map(x->KeplerAList(x,mu),u1)
axis2=map(x->KeplerAList(x,mu),u2)
ecc1=map(x->KeplerEList(x,mu),u1)
ecc2=map(x->KeplerEList(x,mu),u2)

axis1_v,ecc1_v=DataToPlot(axis1,ecc1)
axis2_v,ecc2_v=DataToPlot(axis2,ecc2);
```

```julia
Aerr12=ListRelErr(axis1,axis2)
Eerr12=ListRelErr(ecc1,ecc2)

Aerr12_v,Eerr12_v=DataToPlot(Aerr12,Eerr12);
```

### Semi-Major-Axis

```julia
plot(ti,axis2_v,label=planetslabelsH)
plot!(title=MyTitle,
      xlabel="t",
      ylabel="Semi-Major-Axis",
      xlims=(t0,tend),
)
```

```julia
savefig("./Figures/Semi-Major-Axis.pdf")
```

```julia
xx=map(x->log10.(x),(map(x->abs.(x),Aerr12_v)));
plot(ti,xx,label=planetslabelsH)
plot!(title=MyTitle,
      xlabel="t",
      ylabel="Semi-Major-Axis-Error",
      ylims=(axisya,axisyb),
      xlims=(t0,tend)
)
```

```julia
savefig("./Figures/Semi-Major-Axis-Error.pdf")
```

# Orbits of planets

```julia
# all planets
xlist = map(u -> u[1], qq2)
ylist = map(u -> u[2], qq2)
zlist = map(u -> u[3], qq2)
plt=plot(xlist,ylist,zlist, aspect_ratio=1,label=planetslabels)
for k  in 1:(nbody-1)
  xlist = map(u -> u[k*3+1], qq2)
  ylist = map(u -> u[k*3+2], qq2)
  zlist = map(u -> u[k*3+3], qq2)
    plot!(xlist,ylist,zlist, aspect_ratio=1,label=planetslabels[k+1])
end  
display(plt)
```

```julia
#outer  planets
xbodies=[4,5,6,7,8]
xlist = map(u -> u[1], qq2)
ylist = map(u -> u[2], qq2)
plt=plot(xlist,ylist, aspect_ratio=1,label=planetslabels)
for k  in xbodies
  xlist = map(u -> u[k*3+1], qq2)
  ylist = map(u -> u[k*3+2], qq2)
    plot!(xlist,ylist, aspect_ratio=1,label=planetslabels[k+1])
end  
display(plt)
```

```julia
# Inner planets 
xbodies=[1,2,3,14]
xlist = map(u -> u[1], qq2)
ylist = map(u -> u[2], qq2)
plt=plot(xlist,ylist, aspect_ratio=1,label=planetslabels)
for k  in xbodies
  xlist = map(u -> u[k*3+1], qq2)
  ylist = map(u -> u[k*3+2], qq2)
  plot!(xlist[1:10],ylist[1:10], aspect_ratio=1,label=planetslabels[k+1])
end  
display(plt)
```
