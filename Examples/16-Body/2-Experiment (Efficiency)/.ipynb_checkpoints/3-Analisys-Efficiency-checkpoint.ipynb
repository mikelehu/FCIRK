```julia
#Filenames
#Input

# Eficiency statistics (Threads=1)
StatcsT11="./Data/StacsT11.txt"
StatcsT12="./Data/StacsT12.txt"
StatcsT13="./Data/StacsT13.txt";
StatcsT14="./Data/StacsT14.txt";
# Eficiency statistics (Threads=2)
StatcsT21="./Data/StacsT21.txt"
StatcsT22="./Data/StacsT22.txt"
StatcsT23="./Data/StacsT23.txt";
StatcsT24="./Data/StacsT24.txt";
# Eficiency statistics (Threads=4)
StatcsT41="./Data/StacsT41.txt"
StatcsT42="./Data/StacsT42.txt"
StatcsT43="./Data/StacsT43.txt";
StatcsT44="./Data/StacsT44.txt";
```

<h2 id="Input-statistics">Input-statistics</h2>

```julia
f1=open(StatcsT11,"r");
f2=open(StatcsT12,"r");
f3=open(StatcsT13,"r");
f4=open(StatcsT14,"r");
results1=readdlm(f1)
results2=readdlm(f2)
results3=readdlm(f3)
results4=readdlm(f4)
close(f1)
close(f2)
close(f3)
close(f4)
```

```julia
#f1=open(StatcsT21,"r");
#f2=open(StatcsT22,"r");
f3=open(StatcsT23,"r");
f4=open(StatcsT24,"r");
#results1T2=readcsv(f1)
#results2T2=readcsv(f2)
results3T2=readdlm(f3)
results4T2=readdlm(f4)
#close(f1)
#close(f2)
close(f3)
close(f4)
```

```julia
#f1=open(StatcsT21,"r");
#f2=open(StatcsT22,"r");
f3=open(StatcsT43,"r");
f4=open(StatcsT44,"r");
#results1T2=readcsv(f1)
#results2T2=readcsv(f2)
results3T4=readdlm(f3)
results4T4=readdlm(f4)
#close(f1)
#close(f2)
close(f3)
close(f4)
```

<h2 id="Numerical-Integrations-Threads1">Numerical-Integrations-Threads1</h2>

<h3 id="Data-Threads1">Data-Threads1</h3>

```julia
nk=22*4;
nsuccess1,nsteps1,nfcn1,MaxDE1,CTime1,WTime1=Assing_results(nk,results1)    
nsuccess2,nsteps2,nfcn2,MaxDE2,CTime2,WTime2=Assing_results(nk,results2)    
nsuccess3,nsteps3,nfcn3,MaxDE3,CTime3,WTime3=Assing_results(nk,results3)    
nsuccess4,nsteps4,nfcn4,MaxDE4,CTime4,WTime4=Assing_results(nk,results4)
transpose([nsuccess1,nsuccess2,nsuccess4,nsuccess4])
```

<h3 id="Data-Threads1">Graphics-Threads1</h3>

<h4> Plot (DE/ Nfcn) </h4>

```julia
plot(nfcn1, MaxDE1,  marker = :diamond, label="ABAH1064")
#plot!(nfcn2, MaxDE2, marker = :hexagon,label="CO1035")
plot!(nfcn3, MaxDE3, marker = :circle, label="FCIRK-LQ (s=6)")
plot!(nfcn4, MaxDE4, marker = :square, label="FCIRK-LQ (s=8)")
plot!(title=string(MyTitle," (long)"),
      xlabel="Fcn(log)",
      ylabel="Energy Error(log)",
      ylims=(axisya,axisyb),
#      xlims=(0,tend),
#      yticks=[-1e-16,0,1e-16],
      ytickfont = font(10, "Courier")
)
```
