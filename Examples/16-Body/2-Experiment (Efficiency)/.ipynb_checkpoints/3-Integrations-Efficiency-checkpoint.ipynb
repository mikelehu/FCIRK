<h2 id"Problem-Parameters">Problem-Parameters</h2>

```julia
(nbody,u0,k,Gm,iE,iM)= N16_Barycentric()
pint=convert(Array{Int32},[0]);
```

```julia
dim=3
neq=nbody*2*dim
d = Int(neq/2)
qq=u0[1:d]
vv=u0[d+1:neq];
```

```julia
QQ=ChangeVariableQ(qq,Gm)
VV=ChangeVariableV(vv,Gm)
U0=vcat(QQ[4:end],VV[4:end])
neqH=size(U0)[1]
nbodyH=nbody-1;
```

<a href="#top">Back to the top</a>

<h2 id="Integration-Parameters">Integration-Parameters</h2>

```julia
h=[2.5]
t0=0.
tend=1000.;    # portatilean probatzeko egokia
tspan=(t0,tend);

nout=10;
```

```julia
nk=22*4;
j=1/4;
hh0=zeros(Float64,nk);
for i in 1:nk
#    hh0[i]=j*2^(i/4)
     hh0[i]=j*2^(i/16)
end

hhs8=8/6. *hh0
hhs6=hh0;

#ABAH: denak
#FCIRK:
i0=4*8
i1=nk-4*4;
i1=nk
(i0,i1)
```

```julia
#Integration parameters
n0=round.((tend-t0)./hh0)
ns8=round.((tend-t0)./hhs8)

hhh0=(tend-t0)./n0
hhhs8=(tend-t0)./ns8

saveat0=Integer.(round.((tend-t0)./(nout*hh0)))   
saveats8=Integer.(round.((tend-t0)./(nout*hhs8)))

h0=(tend-t0)./(nout*saveat0)
hs8=(tend-t0)./(nout*saveats8)

nstep0=(tend-t0)./h0
nsteps8=(tend-t0)./hs8

nout0=nstep0./saveat0
nouts8=nsteps8./saveats8;
```

```julia
(h0,[tend], nstep0,nout0,saveat0)
```

<h2 id="Numerical-Integrations-Threads1">Numerical-Integrations-Threads1</h2>

<h3 id="Executions-Theads1">Executions-Theads1</h3>

```julia
now()
```

```julia
threads=1;
results4=zeros(nk,10);

print("FCIRK")
for i in i0:i1
    
    print(i,",");
   
    #FCIRK-LQ  Method (ns=8)     
    ns=8;
    codfun=13; 
    Nkepler=15;
    Moreq=0;
    
    prob=MyODEProblem(Nkepler,Moreq,codfun,tspan,hs8[i],U0,k,Gm,pint);
    res=MySolve(prob,longFCIRK,ns,file4,saveats8[i]);
 
    results4[i,:]=res;
    
end
print("end")
```

```julia
now()
```

<h3 id="Data-Threads1">Data-Threads1</h3>

```julia
f4=open(StatcsT14,"w");
writedlm(f4,results4)
close(f4)
```

```julia
nsuccess4,nsteps4,nfcn4,MaxDE4,CTime4,WTime4=Assing_results(nk,results4)
(nsuccess4)
```
