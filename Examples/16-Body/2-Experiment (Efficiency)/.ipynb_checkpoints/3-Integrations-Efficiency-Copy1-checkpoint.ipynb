<h2 id"Problem-Parameters">Problem-Parameters</h2>

```julia
(nbody,u0,k,Gm,iE,iM)= N16_Barycentric();
pint=convert(Array{Int32},[0]);
```

```julia
dim=3;
neq=nbody*2*dim;
d = Int(neq/2);
qq=u0[1:d];
vv=u0[d+1:neq];
```

```julia
QQ=ChangeVariableQ(qq,Gm);
VV=ChangeVariableV(vv,Gm);
U0=vcat(QQ[4:end],VV[4:end]);
neqH=size(U0)[1];
nbodyH=nbody-1;
```

<a href="#top">Back to the top</a>

<h2 id="Integration-Parameters">Integration-Parameters</h2>

```julia
h=[2.5]
t0=0.
tend=100000.
tend=10000.;  
tend=1000.;    # portatilean probatzeko egokia
tspan=(t0,tend);

nout=10;
```

```julia
nk=22;
j=1/4;
hh0=zeros(Float64,nk);
for i in 1:nk
    hh0[i]=j*2^(i/4)
end

hhs8=8/6. *hh0
#hhs16=(16+8)/6. *hh0;
hhs16=hh0;
```

```julia
#Integration parameters
n0=round.((tend-t0)./hh0);
ns8=round.((tend-t0)./hhs8);
ns16=round.((tend-t0)./hhs16);

hhh0=(tend-t0)./n0;
hhhs8=(tend-t0)./ns8;
hhhs16=(tend-t0)./ns16;

#nout=1;
sampling0=Integer.(round.((tend-t0)./(nout*hh0)));    # Integer ????
samplings8=Integer.(round.((tend-t0)./(nout*hhs8)));
samplings16=Integer.(round.((tend-t0)./(nout*hhs16)));

#Integer(round(2.5))
h0=(tend-t0)./(nout*sampling0);
hs8=(tend-t0)./(nout*samplings8);
hs16=(tend-t0)./(nout*samplings16);
#
nstep0=(tend-t0)./h0;
nsteps8=(tend-t0)./hs8;
nsteps16=(tend-t0)./hs16;
#
nout0=nstep0./sampling0;
nouts8=nsteps8./samplings8;
nouts16=nsteps16./samplings16;
```

<h2 id="Numerical-Integrations-Threads1">Numerical-Integrations-Threads1</h2>

<h3 id="Executions-Theads1">Executions-Theads1</h3>

```julia
threads=1;
results1=zeros(nk,10);
results2=zeros(nk,10);
results3=zeros(nk,10);
results4=zeros(nk,10);

for i in 1:nk
    
    print(i,",");
    
    # ABAH1064 Long
    algorithm=3;
    codmethod=10649;
    codfun=162;

    
    prob=MyODEProblem2(neqH, codmethod, algorithm,codfun,tspan, h0[i],U0,Gm,pint)
    res=MySolve(prob,longComp,file1,sampling0[i]);
    
    results1[i,:]=res;
    
    #CO1035 Long
    algorithm=1
    codmethod=1035;
    codfun=161;
    
    prob=MyODEProblem2(neqH, codmethod, algorithm,codfun,tspan, h0[i],U0,Gm,pint)
    res=MySolve(prob,longComp,file2,sampling0[i]);

    results2[i,:]=res;
    
    #FCIRK-LQ  Method (ns=6)     
    ns=6;
    codfun=13; 
    Nkepler=15;
    Moreq=0;
    
    
    prob=MyODEProblem(Nkepler,Moreq,codfun,tspan,hs16[i],U0,k,Gm,pint);
#    res=MySolve(prob,longFCIRK,ns,file3,samplings16[i]);
    res=MySolve(prob,longFCIRK,ns,file3,samplings16[i]);

    results3[i,:]=res;
    
    #FCIRK-LQ  Method (ns=8)     
    ns=8;
    codfun=13; 
    Nkepler=15;
    Moreq=0;

    
    prob=MyODEProblem(Nkepler,Moreq,codfun,tspan,hs8[i],U0,k,Gm,pint);
#    res=MySolve(prob,longFCIRK,ns,file4,samplings8[i]);
    res=MySolve(prob,longFCIRK,ns,file4,samplings8[i]);
 
    results4[i,:]=res;
    
end
print("end")
```
