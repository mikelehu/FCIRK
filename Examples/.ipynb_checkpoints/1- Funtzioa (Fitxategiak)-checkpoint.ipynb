```julia
nout=501;
neq=90;
myfilename1="../Data/outFCIRCq.bin";
myfilename2="../Data/outC.bin";
myfilename3="../Data/outD.bin";
myfilename4="../Data/outE.bin";
emaitza1=readbin(nout,neq,myfilename1);
emaitza2=readbin(nout,neq,myfilename2);
emaitza3=readbin(nout,neq,myfilename3);
emaitza4=readbin(nout,neq,myfilename4);
```

```julia
mat1=transpose(reshape(emaitza1,neq+1,:));
mat2=transpose(reshape(emaitza2,neq+1,:));
mat3=transpose(reshape(emaitza3,neq+1,:));
mat4=transpose(reshape(emaitza4,neq+1,:));
```

```julia
(rows,cols)=size(mat1);
rows
```

```julia
u1=Vector{Vector{Float64}}(rows);
u2=Vector{Vector{Float64}}(rows);
u3=Vector{Vector{Float64}}(rows);
u4=Vector{Vector{Float64}}(rows);
for i in 1:rows
    u1[i]=mat1[i,:];
    u2[i]=mat2[i,:];
    u3[i]=mat3[i,:];
    u4[i]=mat4[i,:];   
end
```

```julia
nbody=16;
dim=3;
neq=nbody*2*dim;
denborak=mat1[:,1];
nbodyH=nbody-1;
```

```julia
(nbody,u0,Gm,iE,iM)= N16_Cartesian();  #kontuz!!! u0: ondoren ez dut erabiltzen
```

```julia
uu1=map(x->x[2:end],u1);
q1=map(x->x[1:3*nbodyH],uu1);
v1=map(x->x[3*nbodyH+1:end],uu1);
qq1=map(x->AldagaiAldaketaQInv(vcat([0,0,0],x),Gm),q1);
vv1=map(x->AldagaiAldaketaVInv(vcat([0,0,0],x),Gm),v1);
outCar1=[vcat(denborak[1],qq1[1],vv1[1])];
for i in 2:nout
    push!(outCar1,vcat(denborak[i],qq1[i],vv1[i]));
end
#
uu2=map(x->x[2:end],u2);
q2=map(x->x[1:3*nbodyH],uu2);
v2=map(x->x[3*nbodyH+1:end],uu2);
qq2=map(x->AldagaiAldaketaQInv(vcat([0,0,0],x),Gm),q2);
vv2=map(x->AldagaiAldaketaVInv(vcat([0,0,0],x),Gm),v2);
outCar2=[vcat(denborak[1],qq2[1],vv2[1])];
for i in 2:nout
    push!(outCar2,vcat(denborak[i],qq2[i],vv2[i]));
end
#
uu3=map(x->x[2:end],u3);
q3=map(x->x[1:3*nbodyH],uu3);
v3=map(x->x[3*nbodyH+1:end],uu3);
qq3=map(x->AldagaiAldaketaQInv(vcat([0,0,0],x),Gm),q3);
vv3=map(x->AldagaiAldaketaVInv(vcat([0,0,0],x),Gm),v3);
outCar3=[vcat(denborak[1],qq3[1],vv3[1])];
for i in 2:nout
    push!(outCar3,vcat(denborak[i],qq3[i],vv3[i]));
end
#
uu4=map(x->x[2:end],u4);
q4=map(x->x[1:3*nbodyH],uu4);
v4=map(x->x[3*nbodyH+1:end],uu4);
qq4=map(x->AldagaiAldaketaQInv(vcat([0,0,0],x),Gm),q4);
vv4=map(x->AldagaiAldaketaVInv(vcat([0,0,0],x),Gm),v4);
outCar4=[vcat(denborak[1],qq4[1],vv4[1])];
for i in 2:nout
    push!(outCar4,vcat(denborak[i],qq4[i],vv4[i]));
end
```
